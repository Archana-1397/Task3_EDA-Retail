## Bar plot:

```python
sns.set_style("darkgrid")
sns.countplot("Ship Mode",data=data)
```

```python
sns.set_style("darkgrid")
sns.countplot("Segment",data=data,palette='mako')
```

```python
sns.set_style("darkgrid")
sns.countplot("Region",data=data,palette='rocket')
```

```python
sns.set_style("darkgrid")
sns.countplot("Category",data=data,palette='rainbow')
```

```python
#analysing the total number of orders from different states
plt.figure(figsize=(15,10))
data['State'].value_counts().plot.bar()
plt.xlabel("States",size=14)
plt.xticks(size=14)
plt.ylabel("Number of orders",size=10)
plt.show()
```

```python
#Region-wise different sub-categories order count
plt.figure(figsize=(10,6))
sns.countplot(x="Sub-Category",hue="Region",data=data,palette='rainbow')
plt.xticks(rotation="vertical",size=14)
plt.show()
```

## Line plot:

```python
#analysing total profit and sales per sub-category
data.groupby('Sub-Category')['Profit','Sales'].agg(sum).plot.line()
plt.title('Total Profit & Sales per Sub-Category');
```

```python
#Country-wise analysis of quantity, sales, discount and profit
data['State'].value_counts().head(10)
```

```python
data_state=data.groupby(['State'])[['Quantity','Sales','Discount','Profit']].mean()
data_state.head(10)
```

### Pie plot:

```python
#analysis of sales
plt.figure(figsize=(18,18))
data.groupby("State").sum()["Sales"].plot.pie(autopct="%1.1f%%")
plt.title('State_wise analysis of Sales',fontsize=20)
```

```python
#sub-category wise analysis of sale
data_sub_category=data.groupby(['Sub-Category'])[['Quantity','Sales','Discount','Profit']].mean()
data_sub_category.head(10)
```

```python
plt.figure(figsize=(15,15))
data.groupby("Sub-Category").sum()["Sales"].plot.pie(autopct="%1.1f%%")
plt.title('Sub_Category-wise analysis of Sales',fontsize=20)
```

```python
#analysis of profit in different regions
plt.figure(figsize=(10,10))
data.groupby("Region").sum()["Profit"].plot.pie(autopct="%1.1f%%")
plt.title('Region-wise analysis of Profit',fontsize=20)
```

```python
#category-wise analysis of profit
plt.figure(figsize=(10,10))
data.groupby("Category").sum()["Profit"].plot.pie(autopct="%1.1f%%")
plt.title('Category-wise analysis of Profit',fontsize=20)
```
